```python
def duplicados(data_frame):
    # Revisar cuantas filas duplicadas hay 
    duplicados = data_frame.duplicated()
    num_duplicados = duplicados.sum()
    print(f'Hay {num_duplicados} filas duplicadas')
    # Eliminamos y confirmamos que no halla  filas duplicadas
    data_frame = data_frame.drop_duplicates()
    duplicados = data_frame.duplicated()
    num_duplicados = duplicados.sum()
    print(f"Ahora hay {num_duplicados} de filas duplicadas")
    return data_frame
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from stop_words import get_stop_words
# Convertir las palabras a numeros y de cuantas veces estan apareciendo
tfidf = TfidfVectorizer(stop_words="english")

# Crear la matriz TF-IDF de el genero de los animes
# TF-IDF: Evaluar la importancia de una palabra en un documento dentro de una colección de documentos
tfidf_matrix_anime_genre = tfidf.fit_transform(anime_df_clean['genre'])
```

```python
from sklearn.metrics.pairwise import linear_kernel

# Calcular las puntuaciones de similitud de coseno
cosine_sim = linear_kernel(tfidf_matrix_anime_genre,tfidf_matrix_anime_genre)
```

```python
# Obtener como indice los nombres de los animes

indice = pd.Series(anime_df_clean.index, index=anime_df_clean["name"])
for i in range (len(indice)):
    indice.values[i]=i
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # Obtener el índice de el nombre del anime proporcionada como title
    idx=indice[title]
    
    #Esta línea crea una lista de tuplas, donde cada tupla contiene el índice de un anime 
    #y su puntuación de similitud coseno con respecto al anime de referencia 
    sim_score=list(enumerate(cosine_sim[idx]))
    
    #Esta línea ordena la lista sim_score en orden descendente en función de la puntuación de similitud coseno. 
    sim_score = sorted(sim_score, key=lambda x:x[1], reverse=True)
    
    #Esta línea selecciona los elementos de la lista sim_score desde el índice x hasta el índice x.
    sim_score = sim_score[2:7]
    
    #Se extraen los primeros elementos de cada tupla en sim_score y se almacenan en la lista indices_recom.
    indices_recom = [i[0] for i in sim_score]
    #print(anime_df["name"].iloc[indices_recom])
    
    #Esta línea devuelve los nombres de los animes recomendados como un objeto de tipo Series de pandas.
    return anime_df_clean["name"].iloc[indices_recom]
```

# Creando el recomendador de animes dando el nombre de una serie

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Cargar los datos del archivo CSV con información de anime y seroes de television
anime_data = anime_df_clean
tvseries_data = tv_shows_df
```

```python
anime_data.head()
```

```python
tvseries_data.head()
```

```python
tvseries_data = tvseries_data.rename(columns={"title":"name","listed_in":"genre","description":"sypnopsis"})
```

```python
tvseries_data.head()
```

```python
nombre_serie_usuario = input("Ingrese el nombre de una serie de TV: ")
```

```python
nombre_serie_usuario = input("Ingrese el nombre de una serie de TV: ")

title_usuario = tvseries_data[tvseries_data['name'].str.contains(nombre_serie_usuario)].reset_index(drop=True)
title_usuario = title_usuario[title_usuario['name'] == nombre_serie_usuario].head(1)
animes_and_serie = pd.concat([anime_data,title_usuario],axis=0).reset_index(drop=True)

from sklearn.feature_extraction.text import TfidfVectorizer

# Crear la matriz TF-IDF de el genero de los animes
tfidf = TfidfVectorizer(stop_words='english')
animes_and_serie['genre'] = animes_and_serie['genre'].fillna('')  # Rellenar valores nulos
tfidf_matrix_anime = tfidf.fit_transform(animes_and_serie['genre'])

from sklearn.metrics.pairwise import linear_kernel

# Calcular las puntuaciones de similitud de coseno de el genero de los animes
cosine_sim_anime = linear_kernel(tfidf_matrix_anime, tfidf_matrix_anime)


def obtener_recomendaciones(nombre_serie, cosine_sim_anime=cosine_sim_anime, animes_and_serie=animes_and_serie):
    # Obtener el índice de la serie de TV proporcionada por el usuario
    indices = pd.Series(animes_and_serie.index, index=animes_and_serie['name']).drop_duplicates()
    idx = indices[nombre_serie]

    # Calcular las puntuaciones de similitud de la serie con todas las demás series
    sim_scores = list(enumerate(cosine_sim_anime[idx]))

    # Ordenar las puntuaciones de similitud en orden descendente
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Obtener los índices de las 5 series más similares (excluyendo la serie proporcionada por el usuario)
    top_indices = [i[0] for i in sim_scores if i[0] != idx][:10]

    # Obtener los títulos de las series recomendadas
    recomendaciones = animes_and_serie['name'].iloc[top_indices]

    return recomendaciones

# Usamos la funcion
recomendaciones = obtener_recomendaciones(nombre_serie_usuario)

# Imprimir las recomendaciones
print(f"\nRecomendaciones de anime relacionadas con '{nombre_serie_usuario}':")
print(recomendaciones)
```
